```python
from core.read_file import ReadFile
import pandas as pd
import matplotlib.pyplot as plt
```

```python
readFile: ReadFile = ReadFile()
    
if readFile.verify_path() == False:
    readFile.fetch_data('https://www.kaggle.com/api/v1/datasets/download/rohitgrewal/netflix-data')
    
    
data_set:pd.DataFrame = readFile.readCSV()
data_set['Release_Date'] = pd.to_datetime(data_set['Release_Date'], errors="coerce")

data_set['Year'] = data_set['Release_Date'].dt.year

data_set.hist()#para dados numéricos
data_set['Category'].value_counts().plot(kind='bar')
```

```python
cst = data_set[data_set['Title'] == 'House of Cards']

print(cst.Show_Id)
```

```python
releases_per_year = data_set['Year'].value_counts().sort_index()

# Exibir gráfico de barras
plt.figure(figsize=(12,6))
releases_per_year.plot(kind='bar', color='skyblue')
plt.title('Number of TV Shows & Movies Released per Year')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.show()
```
